```python
#create a data
m1=pd.DataFrame({"x": [1, 3, 5], "y": [2, 4, 6]})
m2=pd.DataFrame({"x": [1, 2, 5], "y": [5, 4, 6]})

print ("m1", m1)
print ("m2", m2)
```

```python
#merge  by x
# if a and b then inner join
# if a then left join
# if b then right join
# if a or b then outer join

inner_join = m1.merge(m2, on=["x"], how="inner")
print ("ïnner join", inner_join)


left_join = m1.merge(m2, on=["x"], how="left")
print ("left join", left_join)

right_join = m1.merge(m2, on=["x"], how="right")
print ("right join", right_join)


outer_join = m1.merge(m2, on=["x"], how="outer")
print ("outer join", outer_join)
```
